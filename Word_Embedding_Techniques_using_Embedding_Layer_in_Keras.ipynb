```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent_1=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent_1
```

```python
VOC_SIZE=500
```

```python
onehot_repr=[one_hot(words, VOC_SIZE) for words in sent_1]
print(onehot_repr)
```

# Words Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs = pad_sequences(onehot_repr, padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
model = Sequential()
model.add(Embedding(VOC_SIZE, 10, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
embedded_docs[0]
```

```python
model.predict(embedded_docs[0]).shape
```

```python
model.predict(embedded_docs)
```

```python
embedded_docs[0]
```

```python
model.predict(embedded_docs[0])
```

# Assignment

```python
sent_2=["The world is a better place",
      "Marvel series is my favourite movie",
      "I like DC movies",
      "the cat is eating the food",
      "Tom and Jerry is my favourite movie",
      "Python is my favourite programming language"
      ]
```

```python
from tensorflow.keras.preprocessing.text import one_hot
one_hot_Regr = [one_hot(words, 1000) for words in sent_2]
one_hot_Regr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 10
embedded_docse = pad_sequences(one_hot_Regr, padding='pre', maxlen=sent_length)
embedded_docse
```

```python
model_2 = Sequential()
model_2.add(Embedding(1000, 10, input_length=10))
model_2.compile('adam', 'mse')
```

```python
model_2.predict(embedded_docse[0])
```

```python
model_2.summary()
```

```python
model_2.predict(embedded_docse)
```
